```python
Calculus = int(input("Enter marks of Subject 1: "))
Expository_Writing = int(input("Enter marks of Subject 2: "))
ICT = int(input("Enter marks of Subject 3: "))
Applied_Physics = int(input("Enter marks of Subject 4: "))
Discrete_Structure = int(input("Enter marks of Subject 5: "))

total = Calculus + Expository_Writing + ICT + Applied_Physics + Discrete_Structure
percentage = (total / 500) * 100

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
else:
    grade = "Fail"

print("Total Marks:", total)
print("Percentage:", percentage)
print("Grade:", grade)
```
